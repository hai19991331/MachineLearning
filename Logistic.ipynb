```
def loss(x_data,w_data,y_data):
  y_pre=sigmol(np.dot(x_data,w_data))
  cost1= -y_data * np.log(y_pre)
  cost2= -(1-y_data)*np.log(1-y_pre)
  cost = cost1 + cost2 
  return cost.sum()/len(y_data)
  
```

```
def gradient(x_data,y_data):
  y_pre=sigmol(np.dot(x_data,w_data))
  return np.dot(x_data.T,(y_pre - y_data))
```
